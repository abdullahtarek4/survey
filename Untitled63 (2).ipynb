```python
from scipy.stats import ttest_ind, chi2_contingency

#T-test for numeric features
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.drop('Readiness_Status')
t_test_results = {}

for col in numeric_cols:
    group_ready = df[df['Readiness_Status'] == 1][col]
    group_not_ready = df[df['Readiness_Status'] == 0][col]
    t_stat, p_val = ttest_ind(group_ready, group_not_ready, nan_policy='omit')
    t_test_results[col] = p_val

#p-values under 0.05 (significant)
{col: p for col, p in t_test_results.items() if p < 0.05}
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import GradientBoostingRegressor


from sklearn.metrics import classification_report, accuracy_score
```

```python
log_reg = LogisticRegression(max_iter=1000)
```

```python
rf_clf = RandomForestClassifier(random_state=42)
```

```python
rf_clf.fit(X_train, y_train)
```

```python
rf_preds = rf_clf.predict(X_test)
```

```python
print(" Random Forest:\n", classification_report(y_test, rf_preds))
```

```python
print("\n Accuracy Scores:")
print("Random Forest:", accuracy_score(y_test, rf_preds))
```

```python
catboost_clf = CatBoostClassifier(verbose=0)
```

```python
catboost_clf.fit(X_train, y_train)
```

```python
catboost_preds = catboost_clf.predict(X_test)
```

```python
print("CatBoost:\n", classification_report(y_test, catboost_preds))
```

```python
print("\n Accuracy Scores:")
print("CatBoost:", accuracy_score(y_test, catboost_preds))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
models = {
    "Logistic Regression": log_preds,
    "Random Forest": rf_preds,
    "CatBoost": catboost_preds
}
for name, preds in models.items():
    cm = confusion_matrix(y_test, preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Ready", "Ready"])
    disp.plot(cmap="Blues")
    plt.title(f"Confusion Matrix - {name}")
    plt.show()
```

```python
from sklearn.model_selection import cross_val_score
log_scores = cross_val_score(log_reg, X, y, cv=5)
rf_scores = cross_val_score(rf_clf, X, y, cv=5)
catboost_scores = cross_val_score(catboost_clf, X, y, cv=5)
print("üìò Logistic Regression CV:", log_scores.mean())
print("üå≤ Random Forest CV:", rf_scores.mean())
print("üê± CatBoost CV:", catboost_scores.mean())
```

```python
from sklearn.model_selection import RandomizedSearchCV
param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [4, 6, 10, None],
    'min_samples_split': [2, 5, 10]
}
rf_random = RandomizedSearchCV(rf_clf, param_distributions=param_dist, n_iter=10, cv=3, random_state=42)
rf_random.fit(X_train, y_train)

best_rf = rf_random.best_estimator_

print("üå≤ Best Random Forest Params:", rf_random.best_params_)
print("üå≤ Best RF Accuracy:", accuracy_score(y_test, best_rf.predict(X_test)))
```

## üß† Project Summary

In this project, we aimed to analyze students' readiness for the job market based on academic, personal, and career-related factors using a dataset collected from a student survey.

### üîç Steps We Followed:

1. **Data Cleaning & Preparation:**
   - Checked for nulls, data types, and duplicates.
   - Removed a "score" column that would cause data leakage.
   - Encoded categorical variables and scaled numerical ones.

2. **Descriptive & Inferential Statistics:**
   - Summary statistics (mean, median, std).
   - T-tests and ANOVA to check if specific features influence readiness.

3. **Correlation Analysis:**
   - Pearson heatmap and correlation scores to identify key relationships.

4. **Regression Analysis:**
   - Applied linear regression to explore predictors of readiness score before removing it.

5. **Data Visualization:**
   - Used seaborn/matplotlib to visualize feature distributions, correlations, and relationships.

6. **Feature Selection:**
   - Used two methods: `SelectKBest` and `Random Forest Feature Importances`.

7. **Machine Learning Modeling:**
   - Trained 3 models: Logistic Regression, Random Forest, and CatBoost.
   - Evaluated them using classification reports, accuracy, and confusion matrices.
   - CatBoost gave the best performance overall.

8. **Cross Validation:**
   - Verified model generalization using 5-fold CV.

9. **Hyperparameter Tuning:**
   - Tuned Random Forest using RandomizedSearchCV.

10. **Model Saving and Predictions:**
    - Saved best model using `joblib`.
    - Exported predictions with actual values to CSV.

### üèÅ Conclusion:
We successfully built a predictive model that can classify students as "Ready" or "Not Ready" for the job market with high accuracy. The CatBoost model proved to be the most effective. This project highlights the power of combining statistical analysis with machine learning to derive actionable insights from survey data.
